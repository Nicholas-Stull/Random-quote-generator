````python
import tkinter as tk
import random
import json

# Load quotes from JSON file
with open("quotes.json", "r", encoding="utf-8") as f:
    quotes = json.load(f)

# Create tkinter window
window = tk.Tk()
window.title("Random Quote Generator")

# Create label to display quote
quote_label = tk.Label(window, text="", font=("Helvetica", 18), wraplength=500)
quote_label.pack(pady=20)

# Function to generate and display a random quote


def generate_quote():
    quote = random.choice(quotes)
    quote_text = f'"{quote["quote"]}"\n- {quote["author"]}'
    # Limit quote text to 100 characters
    quote_text = quote_text[:100] + \
        "..." if len(quote_text) > 100 else quote_text
    quote_label.config(text=quote_text)


# Create "Generate Quote" button
generate_button = tk.Button(
    window, text="Generate Quote", command=generate_quote)
generate_button.pack(pady=10)

# Start the tkinter event loop
window.mainloop()
````

In this version of the script, I've added a check to see if the length of the quote text exceeds 100 characters. If it does, the quote text is truncated to 100 characters and an ellipsis is added. If the length of the quote text is 100 characters or less, the full quote text is displayed.

You can adjust the limit of 100 characters to a different value, or you can limit the quote to a specific number of words by using the split() function to split the quote text into words, and then rejoining the first N words using the join() function. For example, to limit the quote to 10 words:
```python
def generate_quote():
    quote = random.choice(quotes)
    quote_text = f'"{quote["quote"]}"\n- {quote["author"]}'
    # Limit quote text to 10 words
    words = quote_text.split()
    quote_text = " ".join(words[:10]) + "..." if len(words) > 10 else quote_text
    quote_label.config(text=quote_text)
```

````python
import tkinter as tk
import random
import json

# Load quotes from JSON file
with open("quotes.json", "r", encoding="utf-8") as f:
    quotes = json.load(f)

# Create tkinter window
window = tk.Tk()
window.title("Random Quote Generator")

# Create label to display quote
quote_label = tk.Label(window, text="", font=("Helvetica", 18), wraplength=500)
quote_label.pack(pady=20)

# Function to generate and display a random quote


def generate_quote():
    quote = random.choice(quotes)
    quote_text = f'"{quote["quote"]}"'
    author_text = f'- {quote["author"]}'
    # Limit quote text to 100 characters
    if len(quote_text) + len(author_text) > 100:
        quote_text = quote_text[:100-len(author_text)-3] + "..."
    quote_label.config(text=f"{quote_text}\n{author_text}")


# Create "Generate Quote" button
generate_button = tk.Button(
    window, text="Generate Quote", command=generate_quote)
generate_button.pack(pady=10)

# Start the tkinter event loop
window.mainloop()
````
